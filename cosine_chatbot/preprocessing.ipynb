```python
 import pandas as pd
```

```python
d=pd.read_csv("x.csv")
```

```python
d.head()
```

```python
inte=[]
que=[]
ans=[]

for i in d['intent,']:
    i=i.split('?')
    
    print(i[0].split(','),i[1])
    ans.append(i[1])
    que.append(i[0].split(',')[1])
    inte.append(i[0].split(',')[0])
```

```python
data =pd.DataFrame(columns =['intent','question','answer'])
```

```python
data['intent']=inte
data['question'] =que
data['answer'] =ans
```

```python
def removing_coma(dat):
    return dat[1:]
data['answer'] =data['answer'].apply(lambda x :removing_coma(x))
```

```python
data
```

```python
data.to_csv('data2.csv',index=False)
```
